#### Handling i/o; reading in data for various sims

```python
# reading in morphological data files for m11b, m11e, and m12i
data_m12i = ascii.read('../data/newruns-m12i-swb_morphology-check.txt')
data_m11e = ascii.read('../data/newruns-m11e-swb_morphology-check.txt')
data_m11b = ascii.read('../data/newruns-m11b-swb_morphology-check.txt')

# array for x axis labels as per sim names
xlabels = ['M+06 DTD', 'M+17 DTD', '$\\alpha=-0.5$ [$10\\times$]', \
           '$\\alpha=-0.5$ [$100\\times$]', '$\\alpha=-1.0$ [$10\\times$]', \
           '$\\alpha=-1.0$ [$100\\times$]']
```

#### Plot: single 4-panel vertical figure comparing stellar mass, R90, v/sigma, and axis ratios for variants of the 3 re-simulations 

```python
# initialising overall figure
fig, ax = plt.subplots(4,1)
fig.set_tight_layout(False)
fig.subplots_adjust(left=0.27, right=0.98, top=0.99, bottom=0.11, wspace=0)

# plotting m12i Mstar values
ax[0].scatter(xlabels, data_m12i['Mgal[Msun]'], marker='o', s=180, c='r', \
           edgecolors='k', linewidths=2, \
           label=None)
# plotting m11e Mstar values
ax[0].scatter(xlabels, data_m11e['Mgal[Msun]'], marker='s', s=160, c='orange', \
           edgecolors='k', linewidths=2, \
           label=None)
# plotting m11b Mstar values
ax[0].scatter(xlabels, data_m11b['Mgal[Msun]'], marker='X', s=200, c='brown', \
           edgecolors='k', linewidths=2, \
           label=None)

# plotting m12i r90 values
ax[1].scatter(xlabels, data_m12i['r90[kpc]'], marker='o', s=180, c='r', \
           edgecolors='k', linewidths=2, \
           label=None)
# plotting m11e r90 values
ax[1].scatter(xlabels, data_m11e['r90[kpc]'], marker='s', s=160, c='orange', \
           edgecolors='k', linewidths=2, \
           label=None)
# plotting m11b r90 values
ax[1].scatter(xlabels, data_m11b['r90[kpc]'], marker='X', s=200, c='brown', \
           edgecolors='k', linewidths=2, \
           label=None)

# plotting m12i v/sigma values
ax[2].scatter(xlabels, data_m12i['v_sigma'], marker='o', s=180, c='r', \
           edgecolors='k', linewidths=2, \
           label=None)
# plotting m11e v/sigma values
ax[2].scatter(xlabels, data_m11e['v_sigma'], marker='s', s=160, c='orange', \
           edgecolors='k', linewidths=2, \
           label=None)
# plotting m11b v/sigma values
ax[2].scatter(xlabels, data_m11b['v_sigma'], marker='X', s=200, c='brown', \
           edgecolors='k', linewidths=2, \
           label=None)

# plotting m12i axis ratio values
ax[3].scatter(xlabels, data_m12i['min_to_maj'], marker='o', s=180, c='r', \
           edgecolors='k', linewidths=2, \
           label='m12i ($\\rm{M_*} \sim 10^{10} \\rm{M}_{\\odot}$)')
# plotting m11e axis ratio values
ax[3].scatter(xlabels, data_m11e['min_to_maj'], marker='s', s=160, c='orange', \
           edgecolors='k', linewidths=2, \
           label='m11e ($\\rm{M_*} \sim 10^{8} \\rm{M}_{\\odot}$)')
# plotting m11b axis ratio values
ax[3].scatter(xlabels, data_m11b['min_to_maj'], marker='X', s=200, c='brown', \
           edgecolors='k', linewidths=2, \
           label='m11b ($\\rm{M_*} \sim 10^{7} \\rm{M}_{\\odot}$)')

# specifying rest of 1st panel
ax[0].set_ylabel('Stellar Mass [$\\rm{M}_{\\odot}$]')
ax[0].set_yscale('log')
ax[0].set_ylim(10**6.1, 10**11.5)
ax[0].set_yticks(ticks=[10**7, 10**8, 10**9, 10**10, 10**11])
ax[0].tick_params(axis='y', which='minor')

# specifying rest of 2nd panel
ax[1].set_ylabel('Stellar $\\rm{R}_{90}$ [kpc]')
ax[1].set_ylim(5, 19)

# specifying rest of 3rd panel
ax[2].set_ylabel("Stellar v/$\\sigma$")
ax[2].set_ylim(0.1, 1.95)

# specifying rest of 4th panel
ax[3].set_ylabel("Axis ratio (min/maj)")
ax[3].set_ylim(0, 0.99)
ax[3].legend(prop={'size':17}, loc='upper left', ncol=1)

plt.xticks(rotation=45, fontsize=16, ha='right')

# eliminate tick labels on common x axes so only bottom panel has x ticks
plt.setp(ax[0].get_xticklabels(), visible=False)
plt.setp(ax[1].get_xticklabels(), visible=False)
plt.setp(ax[2].get_xticklabels(), visible=False)

# specifying rest of overall figure parameter
fig.subplots_adjust(hspace=0)
#plt.savefig('/specify/your/desired/path/here.pdf')
plt.show()
```
