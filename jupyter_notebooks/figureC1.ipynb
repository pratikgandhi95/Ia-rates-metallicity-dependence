#### handling i/o; reading in data along with some manipulations

```python
# reading in DES data shared with us, normalised to 10^10Msun
DES = ascii.read('../data/SNrates_DES_Wiseman21_norm10.txt')
DES_mass = DES['stellarMass']
DES_specRate = DES['specRate']
DES_uperr = DES['specRateErrUp']
DES_lowerr = DES['specRateErrLow']
DES_sigmaUp = np.array(DES_uperr) - np.array(DES_specRate)
DES_sigmaLow = np.array(DES_specRate) - np.array(DES_lowerr)
```

```python
# reading in asas-sn volume-limited data with errors (normalised to Baldry+12 SMF)
asassnVL_B12 = ascii.read('../data/asassn_vollim_mod.txt')
asassnVL_B12_mass = np.array(asassnVL_B12['massBin'])
asassnVL_B12_mass = np.power(10, asassnVL_B12_mass+0.06)
asassnVL_B12_rate = np.array(asassnVL_B12['relRate'])
low_errVL_B12 = np.array(asassnVL_B12['relRateErrLow'])
up_errVL_B12 = np.array(asassnVL_B12['relRateErrUp'])
```

```python
# reading in asas-sn volume-limited upper limits (normalised to Baldry+12 SMF)
asassnVL_B12_uplim = ascii.read('../data/asassn_vollim_mod_uplims.txt')
asassnVL_B12_uplim_mass = np.array(asassnVL_B12_uplim['massBin'])
asassnVL_B12_uplim_mass = np.power(10, asassnVL_B12_uplim_mass+0.06)
asassnVL_B12_uplim_uperr = np.array(asassnVL_B12_uplim['relRateErrUp'])
asassnVL_B12_uplim_uplims = np.array(['1, 1, 1'], dtype='bool')
asassnVL_B12_uplim_yerr = 17
```

```python
# reading in asas-sn volume-limited data with errors (normalised to ZFOURGE SMF)
asassnVL_ZF = ascii.read('../data/asassn_vollim_mod-zfourge.txt')
asassnVL_ZF_mass = np.array(asassnVL_ZF['massBin'])
asassnVL_ZF_mass = np.power(10, asassnVL_ZF_mass-0.06)
asassnVL_ZF_rate = np.array(asassnVL_ZF['relRate'])
low_errVL_ZF = np.array(asassnVL_ZF['relRateErrLow'])
up_errVL_ZF = np.array(asassnVL_ZF['relRateErrUp'])
```

```python
# reading in asas-sn volume-limited upper limits (normalised to ZFOURGE SMF)
asassnVL_ZF_uplim = ascii.read('../data/asassn_vollim_mod-zfourge_uplims.txt')
asassnVL_ZF_uplim_mass = np.array(asassnVL_ZF_uplim['massBin'])
asassnVL_ZF_uplim_mass = np.power(10, asassnVL_ZF_uplim_mass-0.06)
asassnVL_ZF_uplim_uperr = np.array(asassnVL_ZF_uplim['relRateErrUp'])
asassnVL_ZF_uplim_uplims = np.array(['1, 1, 1'], dtype='bool')
asassnVL_ZF_uplim_yerr = 9
```

#### plot 1: comparing all observational samples -- DES, ASAS-SN VL renormalised to Baldry+12 SMF and ASAS-SN VL renormalised to ZFOURGE SMF 

```python
# initialising figure
fig, ax = plt.subplots(1,1)
fig.set_tight_layout(False)
fig.subplots_adjust(left=0.175, right=0.96, top=0.97, bottom=0.15, wspace=0)

# plotting asassn (Baldry+12 SMF) volume-limited data points with error bars (normalised)
ax.scatter(asassnVL_B12_mass, asassnVL_B12_rate, marker='s', s=130, c='k', \
           edgecolors='k', linewidths=2.5, label='ASAS-SN volume-limited (B+12 SMF)', 
           zorder=1000, alpha=1)
ax.errorbar(asassnVL_B12_mass, asassnVL_B12_rate, yerr=(asassnVL_B12_rate-low_errVL_B12,
                                                    up_errVL_B12-asassnVL_B12_rate), 
            fmt='none', ecolor='k', capsize=6.0, capthick=2.0, zorder=1000, 
            label=None, alpha=1)

# plotting asassn (Baldry+12 SMF) volume-limited upper limits
ax.errorbar(asassnVL_B12_uplim_mass, asassnVL_B12_uplim_uperr, 
            yerr=asassnVL_B12_uplim_yerr, uplims=asassnVL_B12_uplim_uplims, fmt='none', 
            ecolor='k', zorder=10, capsize=6.0, capthick=2.0, label=None, alpha=1)

# plotting asassn (ZFOURGE SMF) volume-limited data points with error bars (normalised)
ax.scatter(asassnVL_ZF_mass, asassnVL_ZF_rate, marker='s', s=130, facecolors='none', \
           edgecolors='r', linewidths=2.5, label='ASAS-SN volume-limited (T+14 SMF)', 
           zorder=1000, alpha=0.8)
ax.errorbar(asassnVL_ZF_mass, asassnVL_ZF_rate, yerr=(asassnVL_ZF_rate-low_errVL_ZF,
                                                    up_errVL_ZF-asassnVL_ZF_rate), 
            fmt='none', ecolor='r', capsize=6.0, capthick=2.0, zorder=10000, 
            label=None, alpha=0.8)

# plotting asassn (ZFOURGE SMF) volume-limited upper limits
ax.errorbar(asassnVL_ZF_uplim_mass, asassnVL_ZF_uplim_uperr, 
            yerr=asassnVL_ZF_uplim_yerr, uplims=asassnVL_ZF_uplim_uplims, fmt='none', 
            ecolor='r', zorder=10000, capsize=6.0, capthick=2.0, label=None, alpha=0.8)

# plotting DES sample normalised to 10^10Msun
ax.scatter(DES_mass, DES_specRate, marker='X', s=220, c='blue', linewidths=1.5, 
           edgecolors='k', alpha=1, zorder=100, label='DES (W+21 SMF)')
ax.errorbar(DES_mass, DES_specRate, yerr=(DES_sigmaLow, DES_sigmaUp), fmt='none', 
            ecolor='b', capsize=8.0, alpha=1, capthick=2.5, zorder=100, 
            barsabove=True, label=None)

# specifying rest of the plot
ax.set_xlabel("Stellar Mass [$\\rm{M}_{\odot}$]")
ax.set_ylabel("Relative specific Ia rate")
ax.set_xlim(10**6.8, 10**11.7)
ax.set_ylim(10**-1, 10**3)
ax.legend(prop={'size':18}, loc='upper right', ncol=1)
ax.set_xscale('log')
ax.set_yscale('log')
#plt.savefig('/specify/your/desired/path/here.pdf')
plt.show()
plt.clf()
```
