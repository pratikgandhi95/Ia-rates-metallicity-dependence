```python
# importing necessary packages
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.ticker as t
%matplotlib inline
from astropy.io import ascii
```

```python
# specific matplotlib settings from my matplotlibrc file
plt.rcParams['legend.frameon'] = False
plt.rcParams['savefig.format'] = 'pdf'
plt.rcParams['legend.handletextpad'] = 0.3
plt.rcParams['figure.figsize'] = (8, 7)
plt.rcParams['lines.linewidth'] = 2.5
plt.rcParams['font.family'] = 'serif'
plt.rcParams['font.style'] = 'normal'
plt.rcParams['font.variant'] = 'normal'
plt.rcParams['font.weight'] = 'normal'
plt.rcParams['font.stretch'] = 'normal'
plt.rcParams['font.size'] = 25.0
plt.rcParams['mathtext.fontset'] = 'dejavuserif'
plt.rcParams['axes.linewidth'] = 2.5
plt.rcParams['axes.labelsize'] = 25.0
plt.rcParams['axes.labelpad'] = 6.0
plt.rcParams['axes.titlepad'] = 8.0
plt.rcParams['axes.titlesize'] = 25.0
plt.rcParams['axes.formatter.limits'] = -7,7
plt.rcParams['xtick.major.size'] = 15.0
plt.rcParams['ytick.major.size'] = 15.0
plt.rcParams['xtick.minor.size'] = 8.0
plt.rcParams['ytick.minor.size'] = 8.0
plt.rcParams['xtick.major.pad'] = 12.0
plt.rcParams['ytick.major.pad'] = 12.0
plt.rcParams['xtick.minor.pad'] = 12.0
plt.rcParams['ytick.minor.pad'] = 12.0
plt.rcParams['xtick.major.width'] = 1.6
plt.rcParams['ytick.major.width'] = 1.6
plt.rcParams['xtick.minor.width'] = 0.8
plt.rcParams['ytick.minor.width'] = 0.8
plt.rcParams['xtick.direction'] = 'in'
plt.rcParams['ytick.direction'] = 'in'
plt.rcParams['legend.loc'] = 'best'
plt.rcParams['legend.numpoints'] = 1
plt.rcParams['legend.scatterpoints'] = 1
plt.rcParams['figure.constrained_layout.use'] = False
```

#### handling i/o; reading in data points

```python
# reading in values for Kirby et al. 2013 points
data_K13 = ascii.read('../data/FeAbund_obs_Kirby+13.txt')
stellarMass_K13 = data_K13['stellarMass']
FeAbund_K13 = data_K13['FeAbund']
FeAbund_lowerr_K13 = data_K13['FeAbund_lowerr']
FeAbund_uperr_K13 = data_K13['FeAbund_uperr']
```

```python
# reading in values for Gallazzi et al. 2005 points
data_G05 = ascii.read('../data/FeAbund_obs_Gallazzi+05.txt')
stellarMass_G05 = data_G05['stellarMass']
FeAbund_med_G05 = data_G05['FeAbund_med']
FeAbund_16_G05 = data_G05['FeAbund_16']
FeAbund_84_G05 = data_G05['FeAbund_84']
```

```python
# reading in values for galaxies >10^9Msun (stellarmass-weighted linear mean metallicty)
data_highmass = ascii.read('../data/metallicity_values_sims_final_plot-highmass-SMWLM.txt')
simID_highmass = data_highmass['simID']
stellarMass_highmass = data_highmass['stellarMass']
FeH_highmass = data_highmass['FeH_mass_weighted']
```

```python
# reading in values for galaxies <10^9Msun (mstar-weighted median over all stars)
data_lowmass = ascii.read('../data/metallicity_values_sims_final_plot-lowmass-median-mstar-weighted.txt')
simID_lowmass = data_lowmass['simID']
stellarMass_lowmass = data_lowmass['stellarMass']
FeH_lowmass = data_lowmass['FeAbund_med_W']
```

```python
# reading in values for re-simulations (SMWLM for m12i, mstar-weighted median for m11e/m11b)
data_snia = ascii.read('../data/metallicity_values_sims_newruns-swb_final_plot.txt')
simID_snia = data_snia['simID']
stellarMass_snia = data_snia['stellarMass']
FeH_snia = data_snia['FeH_mass_weighted']
```

#### plot 1: MZR for FIRE-2 sims

```python
# initialising figure
fig, ax = plt.subplots(1,1)
ax2 = ax.twinx()
fig.set_tight_layout(False)
fig.subplots_adjust(left=0.19, right=0.98, top=0.97, bottom=0.15, wspace=0)

# data points from Gallazzi et al. 2005
ax.scatter(stellarMass_G05, FeAbund_med_G05, marker='<', c='darkorchid', alpha=1.0, \
            s=180, edgecolors='black', linewidths=2.5, zorder=10, label='Gallazzi+05 (SDSS)')
ax.errorbar(stellarMass_G05, FeAbund_med_G05, yerr=(FeAbund_16_G05,FeAbund_84_G05), \
             fmt='none', ecolor='darkorchid', alpha=1.0, capsize=6.0, zorder=1, label=None)

# # data points from Kirby et al. 2013
ax.scatter(stellarMass_K13, FeAbund_K13, marker='>', c='blue', alpha=1.0, \
            s=180, edgecolors='black', linewidths=2.5, zorder=10, label='Kirby+13 (Local)')
ax.errorbar(stellarMass_K13, FeAbund_K13, yerr=(FeAbund_lowerr_K13,FeAbund_uperr_K13), \
             fmt='none', ecolor='blue', alpha=1.0, capsize=6.0, zorder=1, label=None)

# plotting FIRE high-mass galaxies
ax2.scatter(stellarMass_highmass, FeH_highmass, c='r', marker='o', s=180, alpha=1.0, \
           edgecolors='k', linewidths=2.5, zorder=100000, label='FIRE-2')

# plotting FIRE low-mass galaxies
ax2.scatter(stellarMass_lowmass, FeH_lowmass, c='r', marker='o', s=180, alpha=1.0, 
            edgecolors='k', linewidths=2.5, zorder=100000, label=None)

# dividing line between low-mass and high-mass
ax.plot([10**9, 10**9], [-5, 5], color='grey', linestyle='dashed', lw=3, 
        zorder=100000000, label=None)

# specifying rest of the plot
ax2.legend(prop={'size':17}, loc='lower right', ncol=1)
ax.legend(prop={'size':17}, loc='upper left', ncol=1)
ax2.get_yaxis().set_visible(False)
ax.set_xlabel("Stellar Mass [$\\rm{M}_{\odot}$]")
ax.set_ylabel("Stellar [Fe/H]")
ax.set_xlim(10**6.1, 10**11.7)
ax.set_ylim(-2.4, 0.4)
ax2.set_ylim(-2.4, 0.4)
ax.set_xscale('log')
plt.savefig("./plots/figure8a_mass-metallicity_obs-sim-OGFIRE.pdf")
plt.show()
plt.clf()
```

#### plot 2: MZR for SN Ia re-simulations

```python
# initialising figure
fig, ax = plt.subplots(1,1)
ax2 = ax.twinx()
fig.set_tight_layout(False)
fig.subplots_adjust(left=0.19, right=0.98, top=0.97, bottom=0.15, wspace=0)

# data points from Gallazzi et al. 2005
ax.scatter(stellarMass_G05, FeAbund_med_G05, marker='<', c='darkorchid', alpha=1, \
            s=180, edgecolors='black', linewidths=2.5, zorder=10, label=None)
ax.errorbar(stellarMass_G05, FeAbund_med_G05, yerr=(FeAbund_16_G05,FeAbund_84_G05), \
             fmt='none', ecolor='darkorchid', alpha=1, zorder=1, capsize=6.0, label=None)

# data points from Kirby et al. 2013
ax.scatter(stellarMass_K13, FeAbund_K13, marker='>', c='blue', alpha=1, \
            s=180, edgecolors='black', linewidths=2.5, zorder=10, label=None)
ax.errorbar(stellarMass_K13, FeAbund_K13, yerr=(FeAbund_lowerr_K13,FeAbund_uperr_K13), \
             fmt='none', ecolor='blue', alpha=1, capsize=6.0, zorder=1, label=None)

# plotting snia FIRE galaxies
for i in range(len(simID_snia)):
    if('m11b' in simID_snia[i]):
        if('Mann' in simID_snia[i]):
            ax.scatter(stellarMass_snia[i], FeH_snia[i], c='brown', marker='o', s=180, \
                       alpha=1, edgecolors='k', linewidths=2.5, zorder=100000)
        elif('Maoz' in simID_snia[i]):
            ax.scatter(stellarMass_snia[i], FeH_snia[i], c='brown', marker='*', s=320, \
                       alpha=1, edgecolors='k', linewidths=2.5, zorder=100000)
        elif('Z05_100x' in simID_snia[i]):
            ax.scatter(stellarMass_snia[i], FeH_snia[i], c='brown', marker='X', s=220, \
                       alpha=1, edgecolors='k', linewidths=2.5, zorder=100000)
        elif('Z05_10x' in simID_snia[i]):
            ax.scatter(stellarMass_snia[i], FeH_snia[i], c='brown', marker='D', s=140, \
                       alpha=1, edgecolors='k', linewidths=2.5, zorder=100000)
        elif('Z1_100x' in simID_snia[i]):
            ax.scatter(stellarMass_snia[i], FeH_snia[i], c='brown', marker='^', s=210, \
                       alpha=1, edgecolors='k', linewidths=2.5, zorder=100000)
        elif('Z1_10x' in simID_snia[i]):
            ax.scatter(stellarMass_snia[i], FeH_snia[i], c='brown', marker='v', s=210, \
                       alpha=1, edgecolors='k', linewidths=2.5, zorder=100000)
    elif('m11e' in simID_snia[i]):
        if('Mann' in simID_snia[i]):
            ax.scatter(stellarMass_snia[i], FeH_snia[i], c='orange', marker='o', s=180, \
                       alpha=1, edgecolors='k', linewidths=2.5, zorder=100000)
        elif('Maoz' in simID_snia[i]):
            ax.scatter(stellarMass_snia[i], FeH_snia[i], c='orange', marker='*', s=320, \
                       alpha=1, edgecolors='k', linewidths=2.5, zorder=100000)
        elif('Z05_100x' in simID_snia[i]):
            ax.scatter(stellarMass_snia[i], FeH_snia[i], c='orange', marker='X', s=220, \
                       alpha=1, edgecolors='k', linewidths=2.5, zorder=100000)
        elif('Z05_10x' in simID_snia[i]):
            ax.scatter(stellarMass_snia[i], FeH_snia[i], c='orange', marker='D', s=140, \
                       alpha=1, edgecolors='k', linewidths=2.5, zorder=100000)
        elif('Z1_100x' in simID_snia[i]):
            ax.scatter(stellarMass_snia[i], FeH_snia[i], c='orange', marker='^', s=210, \
                       alpha=1, edgecolors='k', linewidths=2.5, zorder=100000)
        elif('Z1_10x' in simID_snia[i]):
            ax.scatter(stellarMass_snia[i], FeH_snia[i], c='orange', marker='v', s=210, \
                       alpha=1, edgecolors='k', linewidths=2.5, zorder=100000)
    elif('m12i' in simID_snia[i]):
        if('Mann' in simID_snia[i]):
            ax.scatter(stellarMass_snia[i], FeH_snia[i], c='r', marker='o', s=180, \
                       alpha=1, edgecolors='k', linewidths=2.5, zorder=100000)
        elif('Maoz' in simID_snia[i]):
            ax.scatter(stellarMass_snia[i], FeH_snia[i], c='r', marker='*', s=320, \
                       alpha=1, edgecolors='k', linewidths=2.5, zorder=100000)
        elif('Z05_100x' in simID_snia[i]):
            ax.scatter(stellarMass_snia[i], FeH_snia[i], c='r', marker='X', s=220, \
                       alpha=1, edgecolors='k', linewidths=2.5, zorder=100000)
        elif('Z05_10x' in simID_snia[i]):
            ax.scatter(stellarMass_snia[i], FeH_snia[i], c='r', marker='D', s=140, \
                       alpha=1, edgecolors='k', linewidths=2.5, zorder=100000)
        elif('Z1_100x' in simID_snia[i]):
            ax.scatter(stellarMass_snia[i], FeH_snia[i], c='r', marker='^', s=210, \
                       alpha=1, edgecolors='k', linewidths=2.5, zorder=100000)
        elif('Z1_10x' in simID_snia[i]):
            ax.scatter(stellarMass_snia[i], FeH_snia[i], c='r', marker='v', s=210, \
                       alpha=1, edgecolors='k', linewidths=2.5, zorder=100000)
            
# dividing line between low-mass and high-mass
ax.plot([10**9, 10**9], [-5, 0.2], color='grey', linestyle='dashed', lw=3, 
        zorder=100000000, label=None)
            
# dummy points for legend
ax.scatter(1, 10, c='k', marker='o', s=120, alpha=0.4, edgecolors='k', linewidths=2, \
           label='Mannucci+06 DTD (FIRE-2)')
ax.scatter(1, 10, c='k', marker='*', s=180, alpha=0.4, edgecolors='k', linewidths=2, \
           label='Maoz+17 DTD')
ax2.scatter(1, 10, c='k', marker='D', s=80, alpha=0.4, edgecolors='k', linewidths=2, \
           label='$\\alpha=-0.5$; $f_{\\rm max}=10\\times$')
ax2.scatter(1, 10, c='k', marker='X', s=140, alpha=0.4, edgecolors='k', linewidths=2, \
           label='$\\alpha=-0.5$; $f_{\\rm max}=100\\times$')
ax2.scatter(1, 10, c='k', marker='v', s=130, alpha=0.4, edgecolors='k', linewidths=2, \
           label='$\\alpha=-1.0$; $f_{\\rm max}=10\\times$')
ax2.scatter(1, 10, c='k', marker='^', s=130, alpha=0.4, edgecolors='k', linewidths=2, \
           label='$\\alpha=-1.0$; $f_{\\rm max}=100\\times$')

# specifying rest of the plot
ax.legend(prop={'size':15}, loc='upper left', ncol=1)
ax2.legend(prop={'size':15}, loc='lower right', ncol=1)
ax2.get_yaxis().set_visible(False)
ax.set_xlabel("Stellar Mass [$\\rm{M}_{\odot}$]")
ax.set_ylabel("Stellar [Fe/H]")
ax.set_xlim(10**6.1, 10**11.7)
ax.set_ylim(-2.4, 0.4)
ax2.set_ylim(-2.4, 0.4)
ax.set_xscale('log')
plt.savefig("./plots/figure8b_mass-metallicity_obs-sim-snia.pdf")
plt.show()
plt.clf()
```
