#### Handling i/o; reading in data points along with some manipulations

```python
# reading in all the different data tables for each mass bin
maoz_low = ascii.read('../data/maoz_lbt_ranges_abs_10to6-10to8_updated.txt')
maoz_mid = ascii.read('../data/maoz_lbt_ranges_abs_10to8-10to10_updated.txt')
maoz_high = ascii.read('../data/maoz_lbt_ranges_abs_10to10-10to12_updated.txt')
```

#### Plot: fraction of absolute rates vs lookback time for Maoz+17 model

```python
# initialising figure
fig, ax = plt.subplots(1,1)
fig.set_tight_layout(False)
fig.subplots_adjust(left=0.16, right=0.97, top=0.97, bottom=0.14, wspace=0)

colours = ['black', 'red', 'blue']
lbt = maoz_low['lbt']

# plotting points for low mass bin
plt.plot(lbt, maoz_low['mean'], lw=4, color=colours[0], alpha=1.0, zorder=1, \
         linestyle='solid', label='$\\rm{M_*} \\sim 10^6 - 10^8 \\rm{M}_{\\odot}$')
plt.fill_between(lbt, maoz_low['perc16'], maoz_low['perc84'], color=colours[0], \
                 alpha=0.05, hatch=".", zorder=1, label=None)

# plotting points for medium mass bin
plt.plot(lbt, maoz_mid['mean'], lw=4, color=colours[1], alpha=1.0, zorder=1000, \
         linestyle='solid', label='$\\rm{M_*} \\sim 10^8 - 10^{10} \\rm{M}_{\\odot}$')
plt.fill_between(lbt, maoz_mid['perc16'], maoz_mid['perc84'], color=colours[1], \
                 alpha=0.15, hatch="/", zorder=1000, label=None)

# plotting points for high mass bin
plt.plot(lbt, maoz_high['mean'], lw=4, color=colours[2], alpha=1.0, zorder=100000, \
         linestyle='solid', label='$\\rm{M_*} \\sim 10^{10} - 10^{12} \\rm{M}_{\\odot}$')
plt.fill_between(lbt, maoz_high['perc16'], maoz_high['perc84'], color=colours[2], \
                 alpha=0.25, hatch="\\", zorder=100000, label=None)

# specifying rest of the plot
plt.xticks(np.arange(0, 11, 1))
plt.xlim(0, 10)
plt.ylim(0, 1)
plt.legend(prop={'size': 20}, loc='lower right', ncol=1)
plt.xlabel("X (lookback time range) [Gyr]")
plt.ylabel("Abs Rate(age<X)/Abs Rate(total)")
#plt.savefig('/specify/your/desired/path/here.pdf')
plt.show()
```
