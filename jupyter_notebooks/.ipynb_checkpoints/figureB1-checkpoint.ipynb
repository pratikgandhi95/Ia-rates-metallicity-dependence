#### Plot B1: tau90 comparison

```python
# initialising figure
fig, ax = plt.subplots(1,1)
fig.set_tight_layout(False)
fig.subplots_adjust(left=0.15, right=0.97, top=0.97, bottom=0.15, wspace=0)

# Leitner 2012
ax.plot(stellar_mass_leitner, tau90_leitner, color='dodgerblue', lw=5, 
        alpha=1, zorder=1, label='Leitner12')
ax.fill_between(stellar_mass_leitner, tau90_leitner_lowlim, tau90_leitner_uplim, 
                color='dodgerblue', alpha=0.3, zorder=1)

# Behroozi et al. 2019
ax.plot(stellar_mass_b19_sfg, tau90_b19_sfg, color='navy', lw=5, alpha=1, 
        zorder=1, label='UM (Behroozi+19)')
ax.fill_between(stellar_mass_b19_sfg, tau90_b19_lowlim_sfg, tau90_b19_uplim_sfg, 
                color='navy', zorder=1, alpha=0.3)

# Garrison-Kimmel et al. 2019; all LG dwarves
ax.scatter(lg_mass, lg_tau90, marker='s', s=150, c='blue', edgecolors='black', 
           alpha=0.75, linewidths=2.5, zorder=1, 
           label='Weisz+14a, Skillman+17 (Local Group)')

# original FIRE-2 simulations
ax.scatter(stellar_mass_values_oldruns, tau90_oldruns, marker='o', s=175, 
           alpha=1.0, c='red', edgecolors='k', linewidths=2.5, zorder=100000, 
           label='FIRE-2')

# original FIRE-2 ELVIS
ax.scatter(stellar_mass_values_elvis, tau90_elvis, marker='o', s=175, 
           alpha=1.0, c='red', edgecolors='k', linewidths=2.5, zorder=100000, 
           label=None)

# specifying remainder of the plot
ax.legend(prop={'size':16}, loc='lower left', ncol=1)
ax.set_xlabel("Stellar Mass [$\\rm{M}_{\odot}$]")
ax.set_ylabel('$\\tau_{90}$ [$\\rm{Gyr}$ lookback time from $z=0$]')
ax.set_xscale('log')
ax.set_xlim(10**7, 10**11.7)
ax.set_ylim(7.5, 0)
#plt.savefig('/specify/your/desired/path/here.pdf') 
plt.show()
plt.clf()
```
